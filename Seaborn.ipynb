### La structure d'une fonction seaborn 

sns.fonction(x,y,data,hue,size,style)

x,y,data sont les données à afficher
hue,size,style sont les outils de segmentation

```python
#load un dataset avec seaborn
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.drop(['alone','alive','who','adult_male','embark_town','class'], axis = 1, inplace = True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass', y ='age', data= titanic, hue = 'sex')
```

```python
sns.boxplot(x='pclass', y ='age', data= titanic, hue = 'sex')
```

```python
sns.violinplot(x='pclass', y ='age', data= titanic, hue = 'sex')
```

```python
#regarder la distribution des tarifs
sns.distplot(titanic['fare'])
```

```python
#distribution avec deux variables
sns.jointplot('age','fare', data = titanic, kind = 'hex')
```

```python
sns.heatmap(titanic.corr())
```
