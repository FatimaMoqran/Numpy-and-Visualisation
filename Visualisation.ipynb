```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

### Pyplot

```python
#attention toujours avoir les mêmes dimensions
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#ajouts de paramètres:c : couleurs, lw = épaisseur, ls = style
plt.plot(x,y, c= 'red',lw=4, ls='--')
plt.show()
```

### cycle de vie d'une figure

```python
#début de la figure
plt.figure()
#contenu de la figure
plt.plot(x,y, label= 'quadratique')
plt.plot(x,x**3, label ='cubique')
##on peut ajouter un titre
plt.title('figure1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
#affiche la figure
plt.show()
#enregistrer notre figure
plt.savefig('figure.png') 

#faire une autre figure
plt.figure()
plt.plot(x,y)
```

```python
## subplot grille de graphiques plt.subplot(lignes, colonnes, poisition)
plt.subplot(2,1,1)
plt.plot(x,y,c='red')
plt.subplot(2,1,2)
plt.plot(x,y,c='blue')
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title= 'graphique 1'
plt.subplot(2,2,2)
plt.plot(x,np.sin(x),label='sinus')
plt.plot(x,np.cos(x),label='cosinus')
plt.legend()
plt.show()
```

## façon orientée objet

```python
#permet de creer des figures qui partagent la même absisse
#fig est un objet et ax est un tableau ndarray qui contient des objets
fig, ax = plt.subplots(2,1,sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```
